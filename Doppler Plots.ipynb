```python
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
ramp0 = pd.read_csv("A0000CH1.CSV")
plt.plot(ramp0)
fabryperot = pd.read_csv("A0000CH2.CSV")
plt.plot(fabryperot)
```

```python
fullspectrum1 = pd.read_csv("A0001CH1.CSV")
plt.plot(fullspectrum1)
ramp1 = pd.read_csv("A0001CH2.CSV")
plt.plot(ramp1)

x,y = np.split(fullspectrum1, [-1], 1)
x = np.array(x)
y = np.array(y)
x.flatten()
y.flatten()

def fourgauss_function(x, x1, x2, x3, x4, a1, a2, a3, a4, x01, x02, x03, x04, sigma1, sigma2, sigma3, sigma4):
    return a1*np.exp(-(x1-x01)**2/(2*sigma1**2)) +\
           a2*np.exp(-(x2-x02)**2/(2*sigma2**2)) +\
           a3*np.exp(-(x3-x03)**2/(2*sigma3**2)) +\
           a4*np.exp(-(x4-x04)**2/(2*sigma4**2))

popt, pcov = optimize.curve_fit(fourgauss_function, x, y)
y_fit = fourgauss_function(x, *popt)
chi2 = np.sum((y - y_fit)**2/y)
plt.plot(x, y_fit, label='fit')
```

```python
fullspectrum13 = pd.read_csv("A0013CH1.CSV")
plt.plot(fullspectrum13)
ramp13 = pd.read_csv("A0013CH2.CSV")
plt.plot(ramp13)

x,y = np.split(fullspectrum13, [-1], 1)
x = np.array(x[:2250])
y = np.array(y[:2250])
x = x.flatten()
y = y.flatten()
n = len(x)
mean = sum(x*y)/n
sigma = sum(y*(x-mean)**2)/n

mean = mean.item()
sigma = sigma.item()

def gauss_function(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))

popt, pcov = optimize.curve_fit(gauss_function, x, y, p0 = [1, mean, sigma])
plt.plot(x, gauss_function(x, *popt), label='fit')
```
